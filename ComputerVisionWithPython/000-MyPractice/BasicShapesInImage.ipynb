```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = np.zeros((512,512,3), dtype= np.uint16)
sPoint = (100, 200)
ePoint = (200, 300)
colour = (255, 255, 255)
Thickness = 5
plt.imshow(cv2.rectangle(image, sPoint, ePoint, colour, Thickness))
plt.imshow(cv2.circle(image, tuple((x + 100) for x in sPoint), 100, colour, Thickness))
```
