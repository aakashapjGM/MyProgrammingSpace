```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
def showImg(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")

img = cv.imread("../DATA/crossword.jpg", cv.IMREAD_GRAYSCALE)
showImg(img)

thresValue, binayTh = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
showImg(binayTh)

adaptTh = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 8)
showImg(adaptTh)

blendThresh = cv.addWeighted(src1=binayTh, alpha=0.6, src2=adaptTh, beta=0.4, gamma=0 )
showImg(blendThresh)
```

```python
img = cv.cvtColor(cv.imread("../DATA/bricks.jpg").astype(np.float32)/255, cv.COLOR_BGR2RGB)
gamma = 1/2
gammaImage = np.power(img, gamma)
plt.imshow(gammaImage)
```

```python
def load_img(path):
    img = cv.cvtColor(cv.imread(path).astype(np.float32)/255, cv.COLOR_BGR2RGB)
    return img

def show_img(imgMat):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(imgMat)
```

```python
img = load_img("../DATA/bricks.jpg")
gamma = 1/2
gammaImage = np.power(img, gamma)
show_img(gammaImage)
```
