```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.cvtColor(cv.imread("../DATA/rainbow.jpg"), cv.COLOR_BGR2RGB)
imgGray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
plt.imshow(imgGray, cmap='gray')
```

```python
ret, imgThres = cv.threshold(imgGray, 128, 255,cv.THRESH_BINARY)
plt.imshow(imgThres, cmap='gray')
```
