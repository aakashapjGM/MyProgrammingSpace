```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt 
```

```python
largeImg = cv.cvtColor(cv.imread("../DATA/dog_backpack.png"), cv.COLOR_BGR2RGB)
smallImg = cv.cvtColor(cv.imread("../DATA/watermark_no_copy.png"), cv.COLOR_BGR2RGB)
```

```python
R_smallImg = cv.resize(smallImg, (600, 600))
G_smallImg = cv.cvtColor(R_smallImg, cv.COLOR_RGB2GRAY)
I_smallImg = cv.bitwise_not(G_smallImg)
```

```python
plt.imshow(I_smallImg, cmap='gray')
```

```python
largeImg.shape
```

```python
C_smallImg = cv.cvtColor(I_smallImg, cv.COLOR_GRAY2RGB)
plt.imshow(C_smallImg)
```

```python
plt.imshow(G_smallImg, cmap="gray")
```

```python
M_smallImag = cv.bitwise_and(src1 = R_smallImg, src2 = C_smallImg)
```

```python
plt.imshow(M_smallImag)
```

```python
largeImg.shape
```

```python
plt.imshow(largeImg)
```

```python
rowStart = largeImg.shape[0] - 600
rowEnd = largeImg.shape[0]
colStart = largeImg.shape[1] - 600
colEnd = largeImg.shape[1]
ROI = largeImg[rowStart:rowEnd, colStart:colEnd]
```

```python
plt.imshow(ROI)
```

```python
finalmakedImg = cv.bitwise_or(src1=ROI, src2=M_smallImag)
```

```python
largeImg[rowStart:rowEnd, colStart:colEnd] = finalmakedImg
```

```python
plt.imshow(largeImg)
```
