```python
## Library Import ##
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def showImg(imgMat):
    fig = plt.figure(figsize=(10, 12))
    ax = fig.add_subplot(1, 1, 1)
    ax.imshow(imgMat)
```

```python
## Brick Image Import
rainbow = cv.imread("../DATA/rainbow.jpg")
showBrick = cv.cvtColor(rainbow, cv.COLOR_BGR2RGB)
showImg(showBrick)
```

```python
rainbow.shape[:2]
```

```python
mask = np.zeros(rainbow.shape[:2], np.uint8)
```

```python
plt.imshow(mask, cmap='gray')
```

```python
mask[300:400, 100:400] = 255
```

```python
plt.imshow(mask, cmap="gray")
```

```python
maskedRainbow = cv.bitwise_and(rainbow, rainbow, mask=mask )
```

```python
plt.imshow(maskedRainbow)
```

```python
masked_rainbow_hist = cv.calcHist([rainbow], [2], mask=mask, histSize=[256], ranges=[0, 256])
plt.plot(masked_rainbow_hist)
```

```python
## Importing Gorilla Image 
gorilla = cv.imread("../DATA/gorilla.jpg")
RGB_gorilla = cv.cvtColor(gorilla, cv.COLOR_BGR2RGB)
Gray_gorilla = cv.cvtColor(gorilla, cv.COLOR_BGR2GRAY)
plt.imshow(Gray_gorilla, cmap="gray")
```

```python
Gray_gorilla_hist = cv.calcHist([Gray_gorilla], [0], mask=None, histSize=[255], ranges=[0, 255])
```

```python
plt.plot(Gray_gorilla_hist)
```

```python
eq_gorilla = cv.equalizeHist(Gray_gorilla)
```

```python
plt.imshow(eq_gorilla, cmap='gray')
```

```python
Gray_eq_gorilla_hist = cv.calcHist([eq_gorilla], [0], mask=None, histSize=[255], ranges=[0, 255])
plt.plot(Gray_eq_gorilla_hist)
```
