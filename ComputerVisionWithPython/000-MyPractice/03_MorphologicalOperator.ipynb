```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def loadImg():
    img = np.zeros((600, 600))
    cv2.putText(img=img, text='ABCDE', org=(30, 300), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=5, color=(255, 255, 255), thickness=25, lineType=cv2.LINE_AA )
    return img
```

```python
def showImg(img):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = loadImg()
showImg(img)
```

```python
### Kernel creation for erose ###
img = loadImg()
kernel = np.ones((5, 5))
eroImg = cv2.erode(img, kernel=kernel, iterations=3)
showImg(eroImg)
```

```python
### Loading Image###
img = loadImg()

### Adding Noise on loaded image ###
whiteNoise = np.random.randint(low=0, high=2, size=(600, 600))

### Changing value range from 1 to 255 ###
whiteNoise *= 255

### Merging Created noise with the Image ###
noiseImage = img + whiteNoise

### Creating Kernel and appling Mophological Opening on image ###
kernel = np.ones((5, 5))
morphOpeningImg = cv2.morphologyEx(noiseImage, cv2.MORPH_OPEN, kernel)

showImg(morphOpeningImg)
```

```python
showImg(cv2.dilate(img, kernel=kernel, iterations=3))
```
