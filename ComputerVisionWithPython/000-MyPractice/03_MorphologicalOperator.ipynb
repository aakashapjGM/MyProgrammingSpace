```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def loadImg():
    img = np.zeros((600, 600), np.uint8)
    cv2.putText(img=img, text='ABCDE', org=(30, 300), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=5, color=(255, 255, 255), thickness=25 )
    return img
```

```python
def showImg(img):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = loadImg()
showImg(img)
```

```python
### Kernel creation for erose ###
kernel = np.ones((5, 5), np.uint8)
eroImg = cv2.erode(img, kernel=kernel, iterations=3)
showImg(eroImg)
```

```python
white_noise = (np.random.randint(low=0, high=2, size=(600, 600))) * 255

showImg(white_noise)
```

```python
noiseImg = img + white_noise  
showImg(noiseImg)
```

```python
kernel = np.ones((5, 5), np.uint8)
openImg = cv2.morphologyEx(noiseImg, cv2.MORPH_OPEN, kernel)
```
