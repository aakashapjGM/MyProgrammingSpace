```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math
```

```python
image = np.zeros([512, 512, 3], np.uint8)
```

```python
plt.imshow(image)
```

```python
shapeColor = [255, 255, 255]
shapeThickness = 5
cv2.rectangle(image, [100, 100], [300, 300], shapeColor , shapeThickness)
cv2.circle(image, [200, 200], 100,shapeColor , shapeThickness )
cv2.circle(image, [200, 200], 141,shapeColor , shapeThickness )
cv2.line(image, [100, 100], [300, 300], shapeColor , shapeThickness)
cv2.line(image, [100, 300], [300, 100], shapeColor , shapeThickness)
cv2.line(image, [100, 200], [200, 100], shapeColor , shapeThickness)
cv2.line(image, [200, 100], [300, 200], shapeColor , shapeThickness)
cv2.line(image, [200, 300], [300, 200], shapeColor , shapeThickness)
cv2.line(image, [100, 200], [200, 300], shapeColor , shapeThickness)

myString = "Hello World"

cv2.putText(image, myString, [100, 400], cv2.FONT_HERSHEY_SIMPLEX, 1, shapeColor, shapeThickness - 3)

plt.imshow(image)
```
